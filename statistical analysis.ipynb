```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
dataframe = pd.read_csv('./all_data_monthly.csv')
dataframe.drop(['Index', 'Date', 'Recovered', 'Deceased'], axis=1, inplace=True)
fig, ax = pyplot.subplots(figsize=(8,6))
heatmap = sns.heatmap(dataframe.corr(), vmin=0, vmax=1, annot=True, cmap='YlGnBu')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12});
pyplot.savefig('./analysis/Heatmap.png', dpi=500, bbox_inches='tight')
pyplot.show()
```

```python
# Plot monthly covid cases
dataframe = pd.read_csv('all_data_monthly.csv')
fig, ax = pyplot.subplots(figsize=(4,4))
plot = sns.lineplot(x="Index", y="Confirmed", hue='city', data=dataframe).set_title('Confirmed cases')
fig = plot.get_figure()
fig.savefig("./analysis/Covid.png", bbox_inches='tight')
```
